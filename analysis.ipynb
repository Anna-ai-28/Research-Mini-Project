```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Display data
print("First 5 rows:")
print(df.head())

# Additional insights
print("\nAverage House Value:", df["median_house_value"].mean())
print("Maximum House Value:", df["median_house_value"].max())
print("Minimum House Value:", df["median_house_value"].min())
```

```python
# 1. Bar Graph (Top 10 house prices)
top10 = df.sort_values("median_house_value", ascending=False).head(10)
plt.figure()
plt.bar(range(len(top10)), top10["median_house_value"])
plt.title("Top 10 House Prices")
plt.xlabel("Index")
plt.ylabel("House Value")
plt.show()

# 2. Line Graph (Income vs House Value)
df_sorted = df.sort_values("median_income")
plt.figure()
plt.plot(df_sorted["median_income"], df_sorted["median_house_value"])
plt.title("Income vs House Value")
plt.xlabel("Median Income")
plt.ylabel("House Value")
plt.show()

# 3. Pie Chart (House Price Categories)
df["price_category"] = pd.cut(
    df["median_house_value"],
    bins=[0,100000,300000,500000],
    labels=["Low", "Medium", "High"]
)
plt.figure()
df["price_category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("House Price Distribution")
plt.ylabel("")
plt.show()
```
